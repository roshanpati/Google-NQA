```python
def construct_demo_solution(entry):
    """
    Entry is a dict with keys dict_keys(['annotations', 'document_html', 
    'document_title', 'document_tokens', 'document_url', 'example_id', 
    'long_answer_candidates', 'question_text', 'question_tokens'])
    
    Return format: Return each entry from
    
    {'predictions': [
    {
      'example_id': -2226525965842375672,
      'long_answer': {
        'start_byte': 62657, 'end_byte': 64776,
        'start_token': 391, 'end_token': 604
      },
      'long_answer_score': 13.5,
      'short_answers': [
        {'start_byte': 64206, 'end_byte': 64280,
         'start_token': 555, 'end_token': 560}, ...],
      'short_answers_score': 26.4,
      'yes_no_answer': 'NONE'
    }, ... ]
  }
    """
    dict_return = {}
    dict_return['example_id'] = entry['example_id']
    dict_return['short_answers'] = []
    dict_return['long_answer'] = {}
    dict_return['long_answer']['start_byte'] = entry['annotations'][0]['long_answer']['start_byte']
    dict_return['long_answer']['end_byte'] = entry['annotations'][0]['long_answer']['end_byte']
    dict_return['long_answer']['start_token'] = entry['annotations'][0]['long_answer']['start_token']
    dict_return['long_answer']['end_token'] = entry['annotations'][0]['long_answer']['end_token']
    dict_return['yes_no_answer'] = 'NONE'
    dict_return['long_answer_score'] = 0
    dict_return['short_answers_score'] = 0
    return dict_return
```

```python
def construct_prediction(entries):
    predictions = []
    for entry in entries:
        predictions.append(construct_demo_solution(entry))
    return_dict = {}
    return_dict['predictions'] = predictions
    return return_dict
```

```python
construct_prediction(data)
```

```python
for entry in data:
    total_count += 1
    long_answer_start = entry['annotations'][0]['long_answer']['start_token']
    long_answer_end = entry['annotations'][0]['long_answer']['end_token']
    long_answer_stat.append((long_answer_start, long_answer_end))
    if len(entry['annotations'][0]['short_answers']) == 0:
        no_short_answer += 1
        continue
    short_answer_start = entry['annotations'][0]['short_answers'][0]['start_token']
    short_answer_end = entry['annotations'][0]['short_answers'][0]['end_token']
    if len(entry['annotations'][0]['short_answers']) > 1:
        more_than_one_short_answers += 1
    if short_answer_start >= long_answer_start and short_answer_end <= long_answer_end:
        short_completely_in_long += 1
```
